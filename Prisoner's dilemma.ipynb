# Modelling Prisoner's dilemma

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import choice

%matplotlib inline
```

```python
choices = ['b', 'c']
iterations = 1000
```

```python
# case 1: equal chance of betrayal/cooperation for both prisoners
prob_dist1_1 = [0.5, 0.5]
prob_dist1_2 = [0.5, 0.5]

# case 2: chance of betrayal twice as likely for both prisoners
prob_dist2_1 = [2/3, 1/3]
prob_dist2_2 = [2/3, 1/3]

# case 3: chance of cooeration twice as likely for both prisoners
prob_dist3_1 = [1/3, 2/3]
prob_dist3_2 = [1/3, 2/3]
```

```python
# case 1
p1_1 = []
p2_1 = []

for i in range(iterations):
    choice1 = choice(choices, p=prob_dist1_1)
    choice2 = choice(choices, p=prob_dist1_2)
    if choice1 == 'b' and choice2 == 'b':
        p1_1.append(2)
        p2_1.append(2)
    elif choice1 == 'b' and choice2 == 'c':
        p1_1.append(0)
        p2_1.append(3)
    elif choice1 == 'c' and choice2 == 'b':
        p1_1.append(3)
        p2_1.append(0)
    elif choice1 == 'c' and choice2 == 'c':
        p1_1.append(1)
        p2_1.append(1)
```

```python
# case 2
p1_2 = []
p2_2 = []

for i in range(iterations):
    choice1 = choice(choices, p=prob_dist2_1)
    choice2 = choice(choices, p=prob_dist2_2)
    if choice1 == 'b' and choice2 == 'b':
        p1_2.append(2)
        p2_2.append(2)
    elif choice1 == 'b' and choice2 == 'c':
        p1_2.append(0)
        p2_2.append(3)
    elif choice1 == 'c' and choice2 == 'b':
        p1_2.append(3)
        p2_2.append(0)
    elif choice1 == 'c' and choice2 == 'c':
        p1_2.append(1)
        p2_2.append(1)
```

```python
# case 3
p1_3 = []
p2_3 = []

for i in range(iterations):
    choice1 = choice(choices, p=prob_dist3_1)
    choice2 = choice(choices, p=prob_dist3_2)
    if choice1 == 'b' and choice2 == 'b':
        p1_3.append(2)
        p2_3.append(2)
    elif choice1 == 'b' and choice2 == 'c':
        p1_3.append(0)
        p2_3.append(3)
    elif choice1 == 'c' and choice2 == 'b':
        p1_3.append(3)
        p2_3.append(0)
    elif choice1 == 'c' and choice2 == 'c':
        p1_3.append(1)
        p2_3.append(1)
```

```python
plt.figure(figsize=(10,5))
plt.title('Dilemma of prisoner1\nCumulative prison sentence (' + str(iterations) + ' iterations)')
plt.plot(np.arange(iterations), [sum(p1_1[:i]) for i in range(iterations)], linewidth=4, label='equal chance of betrayal')
plt.plot(np.arange(iterations), [sum(p1_2[:i]) for i in range(iterations)], linewidth=4, label='betrayal twice as likely for both')
plt.plot(np.arange(iterations), [sum(p1_3[:i]) for i in range(iterations)], linewidth=4, label='cooperation twice as likely for both')
#plt.plot(np.arange(iterations), [sum(p1_4[:i]) for i in range(iterations)], linewidth=4, label='prisoner twice as likely to betray')
#plt.plot(np.arange(iterations), [sum(p1_5[:i]) for i in range(iterations)], linewidth=4, label='prisoner twice as likely to cooperate')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Dilemma of prisoner1 and prisoner2\nSum of cumulative prison sentence for both prisoners  (' + str(iterations) + ' iterations)')
plt.plot(np.arange(iterations), np.array([sum(p2_1[:i]) for i in range(iterations)]) + np.array([sum(p1_1[:i]) for i in range(iterations)]), linewidth=4, label='equal chance of betrayal')
plt.plot(np.arange(iterations), np.array([sum(p2_2[:i]) for i in range(iterations)]) + np.array([sum(p1_2[:i]) for i in range(iterations)]), linewidth=4, label='betrayal twice as likely for both')
plt.plot(np.arange(iterations), np.array([sum(p2_3[:i]) for i in range(iterations)]) + np.array([sum(p1_3[:i]) for i in range(iterations)]), linewidth=4, label='cooperation twice as likely for both')

plt.legend()
plt.show()
```

```python
# tit-for-tat strategy
# prisoner2 starts with cooperating
# if prisoner1 cooperates then prisoner2 cooperates in the next iteration
# if prisoner1 betrays then prisoner2 betrays in the next iteration
```

```python
# case 1: prisoner2 twice as likely to cooperate
prob_dist6 = [1/3, 2/3]

# case 2: prisoner2 has equal of betraying and cooperating
prob_dist7 = [0.5, 0.5]

# case 3: prisoner2 twice as likely to betray
prob_dist8 = [2/3, 1/3]
```

```python
# case 1
p1_6 = []
p2_6 = []

tmp = 'c'
for i in range(iterations):
    choice2 = choice(choices, p=prob_dist6)
    choice1 = tmp
    if choice1 == 'b' and choice2 == 'b':
        p1_6.append(2)
        p2_6.append(2)
    elif choice1 == 'b' and choice2 == 'c':
        p1_6.append(0)
        p2_6.append(3)
    elif choice1 == 'c' and choice2 == 'b':
        p1_6.append(3)
        p2_6.append(0)  
    else:
        p1_6.append(1)
        p2_6.append(1)
        
    tmp = choice2
```

```python
# case 2
p1_7 = []
p2_7 = []

tmp = 'c'
for i in range(iterations):
    choice2 = choice(choices, p=prob_dist7)
    choice1 = tmp
    if choice1 == 'b' and choice2 == 'b':
        p1_7.append(2)
        p2_7.append(2)
    elif choice1 == 'b' and choice2 == 'c':
        p1_7.append(0)
        p2_7.append(3)
    elif choice1 == 'c' and choice2 == 'b':
        p1_7.append(3)
        p2_7.append(0)  
    else:
        p1_7.append(1)
        p2_7.append(1)
        
    tmp = choice2
```

```python
# case 3
p1_8 = []
p2_8 = []

tmp = 'c'
for i in range(iterations):
    choice2 = choice(choices, p=prob_dist8)
    choice1 = tmp
    if choice1 == 'b' and choice2 == 'b':
        p1_8.append(2)
        p2_8.append(2)
    elif choice1 == 'b' and choice2 == 'c':
        p1_8.append(0)
        p2_8.append(3)
    elif choice1 == 'c' and choice2 == 'b':
        p1_8.append(3)
        p2_8.append(0)  
    else:
        p1_8.append(1)
        p2_8.append(1)
        
    tmp = choice2
```

```python
plt.figure(figsize=(10,5))
plt.title('Dilemma of prisoner1 using tit-for-tat strategy\nCumulative prison sentence (' + str(iterations) + ' iterations)')
plt.plot(np.arange(iterations), [sum(p1_6[:i]) for i in range(iterations)], linewidth=4, label='prisoner2 twice as likely to betray')
plt.plot(np.arange(iterations), [sum(p1_7[:i]) for i in range(iterations)], linewidth=4, label='prisoner2 equally likely to betray or cooperate')
plt.plot(np.arange(iterations), [sum(p1_8[:i]) for i in range(iterations)], linewidth=4, label='prisoner2 twice as likely to cooperate')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.title('Dilemma of prisoner1 and prisoner2 when prisoner1 employs tit-for-tat\nSum of cumulative prison sentence for both prisoners  (' + str(iterations) + ' iterations)')

plt.plot(np.arange(iterations), np.array([sum(p2_1[:i]) for i in range(iterations)]) + np.array([sum(p1_1[:i]) for i in range(iterations)]), linewidth=4, label='equal chance of betrayal')
plt.plot(np.arange(iterations), np.array([sum(p2_2[:i]) for i in range(iterations)]) + np.array([sum(p1_2[:i]) for i in range(iterations)]), linewidth=4, label='betrayal twice as likely for both')
plt.plot(np.arange(iterations), np.array([sum(p2_3[:i]) for i in range(iterations)]) + np.array([sum(p1_3[:i]) for i in range(iterations)]), linewidth=4, label='cooperation twice as likely for both')

plt.plot(np.arange(iterations), np.array([sum(p2_6[:i]) for i in range(iterations)]) + np.array([sum(p1_6[:i]) for i in range(iterations)]), linewidth=4, label='prisoner2 twice as likely to betray (tit-for-tat)')
plt.plot(np.arange(iterations), np.array([sum(p2_7[:i]) for i in range(iterations)]) + np.array([sum(p1_7[:i]) for i in range(iterations)]), linewidth=4, label='prisoner2 equally likely to betray or cooperate (tit-for-tat)')
plt.plot(np.arange(iterations), np.array([sum(p2_8[:i]) for i in range(iterations)]) + np.array([sum(p1_8[:i]) for i in range(iterations)]), linewidth=4, label='prisoner2 twice as likely to cooperate (tit-for-tat)')

plt.legend()
plt.show()
```

```python
# both prisoners employ tit-for-tat strategy

# case 1: prisoners start off with equal chance of cooperating
# if both betray then increase chance of betrayal
# if both cooperate then increase chance of cooperation
# no asymptotic behaviour in betrayal/cooperation increase

iterations = 1000
trials = 1000

p1_1 = []
p2_1 = []

for i in range(trials):
    p1_dist1 = [0.5, 0.5]
    p2_dist1 = [0.5, 0.5]
    p1_10 = []
    p2_10 = []
    for i in range(iterations):
        choice1 = choice(choices, p=p1_dist1)
        choice2 = choice(choices, p=p2_dist1)
        if choice1 == 'b' and choice2 == 'b':
            p1_10.append(2)
            p2_10.append(2)
            p1_dist1 = [round(((1 - p1_dist1[0])/2), 2), 1-round(((1 - p1_dist1[0])/2), 2)]
            p2_dist1 = [(1 - p2_dist1[0])/2, 1-((1 - p2_dist1[0])/2)]

        elif choice1 == 'b' and choice2 == 'c':
            p1_10.append(0)
            p2_10.append(3)
            p2_dist1 = [round((1 - p2_dist1[0])/2, 2), 1-round((1 - p2_dist1[0])/2, 2)]

        elif choice1 == 'c' and choice2 == 'b':
            p1_10.append(3)
            p2_10.append(0)
            p1_dist1 = [round((1 - p1_dist1[0])/2, 2), 1-round((1 - p1_dist1[0])/2, 2)]

        else:  # both cooperate
            p1_10.append(1)
            p2_10.append(1)
            p1_dist1 = [round((1 - p1_dist1[1])/2, 2), 1-round((1 - p1_dist1[1])/2, 2)]
            p2_dist1 = [round((1 - p2_dist1[1])/2, 2), 1-round((1 - p2_dist1[1])/2, 2)]
            
    p1_1.append(p1_10)
    p2_1.append(p2_10)
```

```python
# both prisoners employ tit-for-tat strategy

# case 2: prisoners start off aggressively
# if both betray then increase chance of betrayal
# if both cooperate then increase chance of cooperation
# no asymptotic behaviour in betrayal/cooperation increase

iterations = 1000
trials = 1000

p1_2 = []
p2_2 = []

for i in range(trials):
    p1_dist1 = [0.9, 0.1]
    p2_dist1 = [0.9, 0.1]
    p1_10 = []
    p2_10 = []
    for i in range(iterations):
        choice1 = choice(choices, p=p1_dist1)
        choice2 = choice(choices, p=p2_dist1)
        if choice1 == 'b' and choice2 == 'b':
            p1_10.append(2)
            p2_10.append(2)
            p1_dist1 = [round(((1 - p1_dist1[0])/2), 2), 1-round(((1 - p1_dist1[0])/2), 2)]
            p2_dist1 = [(1 - p2_dist1[0])/2, 1-((1 - p2_dist1[0])/2)]

        elif choice1 == 'b' and choice2 == 'c':
            p1_10.append(0)
            p2_10.append(3)
            p2_dist1 = [round((1 - p2_dist1[0])/2, 2), 1-round((1 - p2_dist1[0])/2, 2)]

        elif choice1 == 'c' and choice2 == 'b':
            p1_10.append(3)
            p2_10.append(0)
            p1_dist1 = [round((1 - p1_dist1[0])/2, 2), 1-round((1 - p1_dist1[0])/2, 2)]

        else:  # both cooperate
            p1_10.append(1)
            p2_10.append(1)
            p1_dist1 = [round((1 - p1_dist1[1])/2, 2), 1-round((1 - p1_dist1[1])/2, 2)]
            p2_dist1 = [round((1 - p2_dist1[1])/2, 2), 1-round((1 - p2_dist1[1])/2, 2)]
            
    p1_2.append(p1_10)
    p2_2.append(p2_10)
```

```python
# both prisoners employ tit-for-tat strategy

# case 3: prisoners start off passive
# if both betray then increase chance of betrayal
# if both cooperate then increase chance of cooperation
# no asymptotic behaviour in betrayal/cooperation increase

iterations = 1000
trials = 1000

p1_3 = []
p2_3 = []

for i in range(trials):
    p1_dist1 = [0.1, 0.9]
    p2_dist1 = [0.1, 0.9]
    p1_10 = []
    p2_10 = []
    for i in range(iterations):
        choice1 = choice(choices, p=p1_dist1)
        choice2 = choice(choices, p=p2_dist1)
        if choice1 == 'b' and choice2 == 'b':
            p1_10.append(2)
            p2_10.append(2)
            p1_dist1 = [round(((1 - p1_dist1[0])/2), 2), 1-round(((1 - p1_dist1[0])/2), 2)]
            p2_dist1 = [(1 - p2_dist1[0])/2, 1-((1 - p2_dist1[0])/2)]

        elif choice1 == 'b' and choice2 == 'c':
            p1_10.append(0)
            p2_10.append(3)
            p2_dist1 = [round((1 - p2_dist1[0])/2, 2), 1-round((1 - p2_dist1[0])/2, 2)]

        elif choice1 == 'c' and choice2 == 'b':
            p1_10.append(3)
            p2_10.append(0)
            p1_dist1 = [round((1 - p1_dist1[0])/2, 2), 1-round((1 - p1_dist1[0])/2, 2)]

        else:  # both cooperate
            p1_10.append(1)
            p2_10.append(1)
            p1_dist1 = [round((1 - p1_dist1[1])/2, 2), 1-round((1 - p1_dist1[1])/2, 2)]
            p2_dist1 = [round((1 - p2_dist1[1])/2, 2), 1-round((1 - p2_dist1[1])/2, 2)]
            
    p1_3.append(p1_10)
    p2_3.append(p2_10)
```

```python
p1_1_array = np.array(p1_1)
p2_1_array = np.array(p2_1)

p1_2_array = np.array(p1_2)
p2_2_array = np.array(p2_2)

p1_3_array = np.array(p1_3)
p2_3_array = np.array(p2_3)
```

```python
p1_1_avg = np.mean(p1_1_array, axis=0)
p2_1_avg = np.mean(p2_1_array, axis=0)

p1_2_avg = np.mean(p1_2_array, axis=0)
p2_2_avg = np.mean(p2_2_array, axis=0)

p1_3_avg = np.mean(p1_3_array, axis=0)
p2_3_avg = np.mean(p2_3_array, axis=0)
```

```python
plt.figure(figsize=(10,5))

plt.title('Dilemma of prisoner1\nTit-for-tat with probabilities updating\n1000 iterations averaged over 1000 trials')
plt.plot(np.arange(iterations), p1_1_avg, label='case1 (equal probability start)')
plt.plot(np.arange(iterations), p1_2_avg, label='case2 (aggressive start)')
plt.plot(np.arange(iterations), p1_3_avg, label='case3 (passive start)')

plt.show()
```
