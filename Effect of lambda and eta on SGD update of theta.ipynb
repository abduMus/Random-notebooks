# Effect of lambda and eta on SGD update of theta

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# use a single toy example
X = np.array([0,0.5])
y = np.array([1])
```

```python
# initialize theta parameters with constant zero offset
theta = np.array([1,1])
```

```python
def update(y, X, lamb, theta, eta=0.1):
    """
    Updates theta parameter by one step using SGD algorithm
    
    Input:
        y: label of single data point (1x1)
        X: feature vector of single data point (2x1)
        lamb: lambda regularization parameter
        eta: step size
            
    Output:
        Updated value of theta
    """
    return theta + y*X - eta*lamb*np.linalg.norm(theta)
```

```python
eta = 0.01
lambs = [0.99, 0.9, 0.6, 0.4, 0.1, 0.01]
thetas = [update(y, X, i, theta, eta) for i in lambs]

xvals = np.arange(-2, 2, .1)

fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(20)
fig.subplots_adjust(hspace=0.2, wspace=0.1)
for i in range(6):
    ax = fig.add_subplot(2, 3, i+1)
    ax.scatter(X[0],y[0], color ='r', s=80, label='training example')
    ax.plot(xvals, -theta[0]/theta[1]*xvals , 'g', linewidth=2, label='before')
    ax.plot(xvals, (1-theta[0]*xvals)/theta[1] , 'g',linestyle='dashed', linewidth=1)#, label='positive margin before')
    ax.plot(xvals, (-1-theta[0]*xvals)/theta[1] , 'g',linestyle='dashed',linewidth=1)#, label='negative margin before')
    ax.plot(xvals, -thetas[i][0]/thetas[i][1]*xvals , 'b', linewidth=2, label='after')
    ax.plot(xvals, (1-thetas[i][0]*xvals)/thetas[i][1] , 'b',linestyle='dashed', linewidth=1)#, label='positive margin after')
    ax.plot(xvals, (-1-thetas[i][0]*xvals)/thetas[i][1] , 'b',linestyle='dashed',linewidth=1)#, label='negative margin after')
    origin = [0], [0] # origin point
    ax.quiver(*origin, theta[0], theta[1], color='green',scale=8)
    ax.quiver(*origin, thetas[i][0], thetas[i][1], color='blue',scale=8)
    ax.legend(loc='upper right')
    ax.grid()
    plt.xlim(-2,2)
    plt.ylim(-2,2)
    plt.title('lambda = ' + str(lambs[i]) + ' eta = ' + str(eta))
plt.show()   
```
