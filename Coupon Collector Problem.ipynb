```python
import numpy as np
import random
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
ks = np.arange(1,1000, 100)
iterations = 1000
result = []

for k in ks:
    temp = []
    choices = np.arange(k)
    n = len(choices)
    for i in range(iterations):
        bag = []
        count = 0
        while len(bag) < n:
            choice = random.choice(choices)
            if choice not in bag:
                bag.append(choice)
            count += 1
        temp.append(count)
    result.append(np.mean(temp))
```

```python
plt.figure(figsize=(10,5))
plt.plot(ks, result, 'r')
plt.plot(ks, [k*math.log(k,2.5) for k in ks], 'g')
plt.show()
```
